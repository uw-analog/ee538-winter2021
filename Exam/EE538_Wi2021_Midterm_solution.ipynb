Assuming all transistors are in saturation, use the small-signal model to answer the following in terms of $g_{m1}$,$g_{m2}$, $g_{m3}$, $g_{m4}$, $r_{o1}$,$r_{o2}$,$r_{o3}$,$r_{o4}$ and $A_0$. 

The gain factor $-A_0$ is the small-signal voltage gain between the input and output terminals of an ideal voltage amplifier (i.e. $v_{g2} = -A_0\cdot v_{s2}$ and $v_{g3} = -A_0\cdot v_{s3}$).

Ignore the body effect (i.e. $\gamma = 0$ for all devices). Justify any assumptions.

__a)__ What is the small-signal output resistance of the circuit, $R_o$?

The output resistance can be determined by breaking the small-signal circuit model into symmetrical halves about the node $v_{out}$. Based on the relation $v_{g2,3} = -A_0 v_{s2,3}$, the trasconductance elements of $M_2$ and $M_3$ are controlled by a small-signal voltage $v_{gs} = (A_0+1)v_s$. The equivalent half-circuit is shown below.

The circuit can be described by a KCL equation at node $v_t$ and Ohm's law for $r_{o1,4}$. For simplicity, $v_s = v_{s2,3}$, $g_m = g_{m2,3}$, and $r_o = r_{o2,3} = r_{o1,4}$.

\begin{equation}
i_t = g_{m}(A_0+1)v_{s} + \dfrac{v_t - v_s}{r_o}
\end{equation}

\begin{equation}
i_t = \dfrac{v_s}{r_o}
\end{equation}

Substituting $v_s = i_t \cdot r_o$ into the KCL expression and solving for $v_t/i_t$ gives

\begin{equation}
\boxed{\dfrac{v_t}{i_t} = 2r_o + g_m r_o^2(A_0+1)}
\end{equation}

The output impedance of the amplifier is the parallel combination of $R_{up}$ and $R_{dn}$:

\begin{equation}
\boxed{R_o = R_{up}||R_{dn} = \left[ r_{o1}+r_{o2} + g_{m2} r_{o1}r_{o2}(A_0+1) \right] || \left[ r_{o3}+r_{o4} + g_{m3} r_{o3}r_{o4}(A_0+1) \right]  }
\end{equation}

We can simplify this expression by noting that $g_m r_o^2 A_0 >> r_o, 1$:

\begin{equation}
R_o \approx \left[ g_{m2} r_{o1}r_{o2}A_0 \right] || \left[ g_{m3}r_{o3}r_{o4}A_0 \right] 
\end{equation}

__b)__ What is the small-signal gain of the circuit, $v_{out}/v_{in}$?

To determine the small-signal gain, we need to find the equivalent transconductance $G_m$. One way of looking at this is that the transconductance-based current generated by $v_{in}$ is subject to a current divider comprised of $M_1$'s output resistance and the resistance looking into $M_2$'s source ($R_{up}$ in the figure below).

The current division exprssion is given by

\begin{equation}
i_{out} = -g_{m1}v_{in} \dfrac{r_{o1}}{r_{o1} + R_{up}}
\end{equation}

$R_{up}$ is the parallel combination of $r_{o2}$ and the equivalent impedance of $M_2$'s transconductance element. Noting that the ground connection at $M_2$'s drain makes the transconductance current wholly dependent on $v_s$, the equivalent impedance can be written as $1/[g_{m2}(A_0+1)]$. $R_{up}$ is thus

\begin{equation}
R_{up} = \dfrac{r_{o2}}{g_{m2}(A_0+1)r_{o2} + 1}
\end{equation}

The expression for $i_{out}$ then becomes

\begin{equation}
i_{out} = -g_{m1}v_{in} \dfrac{r_{o1}[g_{m2}(A_0+1)r_{o2} + 1]}{r_{o1}[g_{m2}(A_0+1)r_{o2} + 1] + r_{o2}}
\end{equation}

The equivalent transconductance is thus

\begin{equation}
\boxed{G_m = -\dfrac{i_{out}}{v_{in}} = g_{m1} \dfrac{r_{o1}[g_{m2}(A_0+1)r_{o2} + 1]}{r_{o1}[g_{m2}(A_0+1)r_{o2} + 1] + r_{o2}} \approx g_{m1} }
\end{equation}

where the approximation is based on the assumption $r_{o1}[g_{m2}(A_0+1)r_{o2} + 1] >> r_{o2}$. 

Finally, the voltage gain of the amplifier is given by

\begin{equation}
\boxed{ \dfrac{v_{out}}{v_{in}} = -G_m R_o \approx -g_{m1} \cdot \left[ g_{m2} r_{o1}r_{o2}A_0 \right] || \left[ g_{m3}r_{o3}r_{o4}A_0 \right] }
\end{equation}
